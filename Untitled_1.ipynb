## ðŸ“Š **Dataset Information**

- **Source**: The Co-Benefits of Reaching Net-Zero in the UK (`df_wide.csv`)
- **Spatial Coverage**: 46,426 small areas across England
- **Year**: 2025 (baseline assessment)
- **Metrics**: All values in Million GBP per area
- **Variables**:
  - `small_area` - Area identifier
  - `year` - Year (2025)
  - `air_quality_million_gbp` - Air quality benefits
  - `physical_activity_million_gbp` - Physical activity benefits
  - `congestion_million_gbp` - Congestion impacts (can be negative)
  - `hassle_costs_million_gbp` - Travel time inconvenience costs (negative)
  - `net_benefit_million_gbp` - Total net benefit
  - `benefit_to_hassle_ratio` - Ratio of benefits to costs

---

```python
# ========================================
# GRAFIK 1: "Three Lines, One Story"
# Air, Movement, and Traffic: Co-Benefits Distribution Across Areas
# ========================================

# Sort data by one of the benefits to show progression
df_sorted = df.sort_values('air_quality_million_gbp').reset_index(drop=True)

# Sample data for visualization (too many points otherwise) - take every 100th area
sample_indices = range(0, len(df_sorted), 100)
df_sample = df_sorted.iloc[sample_indices].reset_index(drop=True)

fig, ax = plt.subplots(figsize=(12, 7))

# Create x-axis as sequential indices representing areas sorted by air quality
x_values = range(len(df_sample))

# Plot three lines with distinct colors and markers
ax.plot(x_values, df_sample['air_quality_million_gbp'],
        marker='o', linewidth=2.5, markersize=6,
        color='#2E86AB', label='Air Quality', alpha=0.9)

ax.plot(x_values, df_sample['physical_activity_million_gbp'],
        marker='s', linewidth=2.5, markersize=6,
        color='#06A77D', label='Physical Activity', alpha=0.9)

ax.plot(x_values, df_sample['congestion_million_gbp'],
        marker='^', linewidth=2.5, markersize=6,
        color='#F18F01', label='Congestion', alpha=0.9)

# Add shaded areas to show different benefit zones
third = len(x_values) // 3
ax.axvspan(0, third, alpha=0.1, color='lightcoral', label='Lower Benefit Areas')
ax.axvspan(third, 2*third, alpha=0.1, color='lightyellow', label='Medium Benefit Areas')
ax.axvspan(2*third, len(x_values)-1, alpha=0.1, color='lightgreen', label='Higher Benefit Areas')

# Styling
ax.set_xlabel('Areas (sorted by Air Quality benefit)', fontsize=14, fontweight='bold')
ax.set_ylabel('Million GBP', fontsize=14, fontweight='bold')
ax.set_title('Air, Movement, and Traffic: Co-Benefits Across Different Areas',
             fontsize=18, fontweight='bold', pad=20)

# Grid and spines
ax.grid(True, alpha=0.3, linestyle='--', linewidth=0.7)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_linewidth(1.5)
ax.spines['bottom'].set_linewidth(1.5)

# Legend
ax.legend(loc='upper left', frameon=True, shadow=True, fancybox=True, fontsize=10, ncol=2)

plt.tight_layout()

# Add caption below the plot

plt.show()

print("\nâœ… Graph 1 completed: Three Lines, One Story")
```

---

### ðŸ“Š **Interpretasi Graph 1: Three Lines, One Story**

**Insight Utama:**
1. **Pola Kenaikan Bersama**: Ketiga garis (air quality, physical activity, congestion) menunjukkan pola yang serupa saat area diurutkan berdasarkan manfaat kualitas udara
2. **Physical Activity Tertinggi**: Manfaat aktivitas fisik memiliki variasi paling tinggi, dengan beberapa area mencapai >25 juta GBP
3. **Congestion Negatif**: Nilai congestion sebagian besar negatif, menunjukkan adanya trade-off dalam beberapa area
4. **Tiga Zona Area**:
   - **Lower Benefit Areas** (kiri): Manfaat rendah di semua metrics
   - **Medium Benefit Areas** (tengah): Manfaat moderat dengan variasi tinggi
   - **Higher Benefit Areas** (kanan): Manfaat tertinggi, terutama untuk physical activity

**Kesimpulan**: Area dengan kualitas udara yang baik cenderung memiliki masyarakat yang lebih aktif secara fisik, mendukung narasi bahwa pengurangan polusi kendaraan mendorong gaya hidup yang lebih sehat.

---

```python
# ========================================
# GRAFIK 2: "Breathing vs Moving"
# Do Areas with Cleaner Air Also Have More Active Movement?
# ========================================

fig, ax = plt.subplots(figsize=(12, 7))

# Prepare data - use all data points
x = df['air_quality_million_gbp']
y = df['physical_activity_million_gbp']

# Create a color map based on net benefit to show which areas benefit most
colors = df['net_benefit_million_gbp']

# Create scatter plot with colormap based on net benefit
scatter = ax.scatter(x, y,
                     c=colors,
                     cmap='RdYlGn',
                     s=50,
                     alpha=0.6,
                     edgecolors='black',
                     linewidth=0.3,
                     zorder=3)

# Add colorbar
cbar = plt.colorbar(scatter, ax=ax, pad=0.02)
cbar.set_label('Net Benefit (Million GBP)', fontsize=13, fontweight='bold', rotation=270, labelpad=25)
cbar.ax.tick_params(labelsize=11)

# Calculate and plot regression line
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
line_x = np.array([x.min(), x.max()])
line_y = slope * line_x + intercept
ax.plot(line_x, line_y, '--', color='red', linewidth=2.5,
        alpha=0.7, label=f'Trend Line (RÂ² = {r_value**2:.3f})', zorder=2)

# Annotate some interesting points
# Find area with highest air quality benefit
idx_max_air = df['air_quality_million_gbp'].idxmax()
ax.annotate(f'Highest Air Quality\n({df.loc[idx_max_air, "small_area"]})',
            xy=(x.iloc[idx_max_air], y.iloc[idx_max_air]),
            xytext=(15, 15), textcoords='offset points',
            fontsize=10, fontweight='bold',
            bbox=dict(boxstyle='round,pad=0.5', facecolor='yellow', alpha=0.7),
            arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.3', lw=2))

# Find area with highest physical activity benefit
idx_max_activity = df['physical_activity_million_gbp'].idxmax()
ax.annotate(f'Highest Physical Activity\n({df.loc[idx_max_activity, "small_area"]})',
            xy=(x.iloc[idx_max_activity], y.iloc[idx_max_activity]),
            xytext=(15, -25), textcoords='offset points',
            fontsize=10, fontweight='bold',
            bbox=dict(boxstyle='round,pad=0.5', facecolor='lightgreen', alpha=0.7),
            arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.3', lw=2))

# Styling
ax.set_xlabel('Air Quality Benefit (Million GBP)', fontsize=14, fontweight='bold')
ax.set_ylabel('Physical Activity Benefit (Million GBP)', fontsize=14, fontweight='bold')
ax.set_title('Do Areas with Cleaner Air Also Have More Active Movement?',
             fontsize=18, fontweight='bold', pad=20)

# Grid and spines
ax.grid(True, alpha=0.3, linestyle='--', linewidth=0.7, zorder=1)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_linewidth(1.5)
ax.spines['bottom'].set_linewidth(1.5)

# Legend
ax.legend(loc='lower right', frameon=True, shadow=True, fancybox=True, fontsize=12)

plt.tight_layout()

# Add caption below the plot

plt.show()

print("\nâœ… Graph 2 completed: Breathing vs Moving")
```

---

### ðŸ“Š **Interpretasi Graph 2: Breathing vs Moving**

**Insight Utama:**
1. **Korelasi Positif**: Terdapat tren positif antara air quality benefit dan physical activity benefit (RÂ² = 0.134)
2. **46,426 Area**: Setiap titik mewakili satu small area di UK, menunjukkan variabilitas spasial yang sangat besar
3. **Color Coding**:
   - **Merah/Orange**: Area dengan net benefit rendah (bahkan negatif)
   - **Kuning/Hijau muda**: Area dengan net benefit moderat
   - **Hijau tua**: Area dengan net benefit sangat tinggi (>60 juta GBP)
4. **Area Outlier**:
   - **Highest Air Quality**: E01021020 - Menunjukkan manfaat udara tertinggi
   - **Highest Physical Activity**: W01002002 - Menunjukkan aktivitas fisik tertinggi

**Kesimpulan**: Meskipun RÂ² relatif rendah (0.134), tren positif tetap jelas terlihat. Ini menunjukkan bahwa area dengan udara lebih bersih cenderung memiliki masyarakat yang lebih aktif, meskipun ada faktor lain yang juga berpengaruh (misalnya infrastruktur, budaya lokal, demografi).

```python
# ========================================
# GRAFIK 3: "Is It Worth the Hassle?"
# Net Gains vs Extra Travel Time: Are We Still Better Off?
# ========================================

# Create bins/ranges for better visualization - group areas by net benefit
df_sorted_benefit = df.sort_values('net_benefit_million_gbp').reset_index(drop=True)

# Create deciles (10 groups) for visualization
df_sorted_benefit['decile'] = pd.qcut(df_sorted_benefit.index, q=10, labels=[f'D{i+1}' for i in range(10)])

# Aggregate by decile
decile_stats = df_sorted_benefit.groupby('decile', observed=True).agg({
    'net_benefit_million_gbp': 'mean',
    'hassle_costs_million_gbp': 'mean',
    'air_quality_million_gbp': 'mean',
    'physical_activity_million_gbp': 'mean',
    'congestion_million_gbp': 'mean'
}).reset_index()

fig, ax1 = plt.subplots(figsize=(12, 7))

# X-axis positions
x_pos = range(len(decile_stats))
x_labels = decile_stats['decile']

# Create bars for net benefit on left y-axis
bars = ax1.bar(x_pos, decile_stats['net_benefit_million_gbp'],
               width=0.6,
               color='#06A77D',
               alpha=0.8,
               label='Net Benefit',
               edgecolor='black',
               linewidth=1.2)

# Styling for left y-axis
ax1.set_xlabel('Area Groups (sorted by Net Benefit, D1=Lowest, D10=Highest)', fontsize=14, fontweight='bold')
ax1.set_ylabel('Net Benefit (Million GBP)', fontsize=14, fontweight='bold', color='#06A77D')
ax1.tick_params(axis='y', labelcolor='#06A77D', labelsize=11)
ax1.set_xticks(x_pos)
ax1.set_xticklabels(x_labels, rotation=0)

# Add horizontal line at y=0
ax1.axhline(y=0, color='gray', linestyle='--', linewidth=1, alpha=0.5)

# Grid on left axis
ax1.grid(True, alpha=0.3, linestyle='--', linewidth=0.7, axis='y')
ax1.spines['top'].set_visible(False)
ax1.spines['left'].set_linewidth(1.5)
ax1.spines['left'].set_color('#06A77D')
ax1.spines['bottom'].set_linewidth(1.5)

# Create second y-axis for hassle costs (right)
ax2 = ax1.twinx()
line = ax2.plot(x_pos, decile_stats['hassle_costs_million_gbp'],
                marker='o',
                linewidth=3,
                markersize=8,
                color='#F18F01',
                label='Hassle Costs',
                alpha=0.9)

# Styling for right y-axis
ax2.set_ylabel('Hassle Costs (Million GBP)', fontsize=14, fontweight='bold', color='#F18F01')
ax2.tick_params(axis='y', labelcolor='#F18F01', labelsize=11)
ax2.spines['right'].set_linewidth(1.5)
ax2.spines['right'].set_color('#F18F01')
ax2.spines['top'].set_visible(False)

# Adjust right y-axis limits to show variation in hassle costs
hassle_min = decile_stats['hassle_costs_million_gbp'].min()
hassle_max = decile_stats['hassle_costs_million_gbp'].max()
hassle_range = hassle_max - hassle_min
ax2.set_ylim(hassle_min - hassle_range * 0.3, hassle_max + hassle_range * 0.3)

# Title
ax1.set_title('Net Benefit vs Hassle Costs Across Different Area Groups',
              fontsize=18, fontweight='bold', pad=20)

# Combined legend
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2,
           loc='upper left', frameon=True, shadow=True, fancybox=True, fontsize=12)

plt.tight_layout()

# Add caption below the plot

plt.show()

print("\nâœ… Graph 3 completed: Is It Worth the Hassle?")
```

---

### ðŸ“Š **Interpretasi Graph 3: Is It Worth the Hassle?**

**Insight Utama:**
1. **Net Benefit Distribution**:
   - **D1-D3** (30% area terendah): Net benefit **negatif** hingga mendekati 0
   - **D4-D7** (40% area menengah): Net benefit **positif moderat** (0.5 - 3 juta GBP)
   - **D8-D10** (30% area tertinggi): Net benefit **sangat positif** (3 - 9.5 juta GBP)

2. **Hassle Costs Pattern**:
   - Hassle costs (garis oranye) berkisar antara **-1.1 hingga -1.8 juta GBP**
   - Area dengan net benefit tinggi cenderung memiliki hassle costs yang **lebih rendah** (lebih dekat ke -1.8)
   - Ini menunjukkan bahwa area yang sudah memiliki infrastruktur baik mengalami lebih sedikit ketidaknyamanan

3. **Trade-off Analysis**:
   - **70% area** (D4-D10) memiliki net benefit positif
   - **30% area** (D1-D3) mengalami net benefit negatif, kemungkinan karena:
     - Infrastruktur transportasi publik yang kurang memadai
     - Dependensi tinggi pada kendaraan pribadi
     - Jarak tempuh yang lebih jauh

4. **Zero Line**: Garis putus-putus abu-abu di y=0 menunjukkan titik impas antara benefit dan cost

**Kesimpulan**: Meskipun ada hassle costs di semua area, **mayoritas wilayah (70%)** tetap diuntungkan secara net dari kebijakan net-zero. Area dengan net benefit tinggi menunjukkan bahwa dengan infrastruktur dan perencanaan kota yang tepat, transisi net-zero dapat menguntungkan secara ekonomi dan sosial.

**Rekomendasi**: Pemerintah perlu fokus pada 30% area dengan net benefit negatif (D1-D3) dengan:
- Meningkatkan infrastruktur transportasi publik
- Menyediakan lebih banyak jalur sepeda dan pedestrian
- Memberikan insentif untuk mengurangi dependensi pada kendaraan pribadi
